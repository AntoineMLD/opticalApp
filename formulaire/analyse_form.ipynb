```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('questionnaire.csv')
df.head()
```

```python
df.columns

```

```python
# Extraire le nom de la première colonne et créer une liste pour les autres questions
nouvelles_colonnes = ['Horodateur'] + [f'Question {i+1}' for i in range(1, len(df.columns))]

# Renommer les colonnes
df.columns = nouvelles_colonnes
```

```python
df.head()
```

```python
question1_counts = df['Question 2'].value_counts()

# Création du diagramme en barres
plt.figure(figsize=(10, 6))
sns.barplot(x=question1_counts.values, y=question1_counts.index)
plt.title('retrouver les informations sur les verres gravés')
plt.xlabel('réponses')
plt.xticks(rotation = 90)
plt.ylabel('Fréquence')
plt.show()
```

```python
question3_counts = df['Question 3'].value_counts()

# Création du diagramme en barres
plt.figure(figsize=(10, 6))
sns.barplot(x=question3_counts.values, y=question3_counts.index)
plt.title('temps consacré à la recherche')
plt.xlabel('Opticiens')
plt.xticks(rotation = 90)
plt.ylabel('Fréquence')
plt.show()
```

```python
question5_counts = df['Question 5'].value_counts()

# Création du diagramme en barres
plt.figure(figsize=(10, 6))
sns.barplot(x=question5_counts.values, y=question5_counts.index)
plt.title('Seriez-vous intéressé par une solution numérique qui permettrait d’accélérer ce processus ?')
plt.xlabel('réponses')
plt.xticks(rotation = 90)
plt.ylabel('Fréquence')
plt.show()
```
